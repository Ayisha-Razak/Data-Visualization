# Example1 – Overlapped Area Plot:

```python
import pandas as pd

import matplotlib.pyplot as plot

# Peak Temperature data for two cities

tempData    = {"Vancouver":[99, 106, 102, 78],

               "Victoria":[77, 84, 80, 85]}
#index
seasons     = ("Spring", "Summer", "Fall", "Winter")


dataFrame   = pd.DataFrame(tempData, index=seasons)
 

#Draw an area plot for the DataFrame data

dataFrame.plot(kind='area', stacked=False)
plot.title('Overlapped area plot: Peak Temperature For Vancouver and Victoria in all seasons')
plot.ylabel('Peak Temperature')
plot.xlabel('Seasons')

plot.show()
```

# Example1 – Stacked Area Plot:

```python

# Draw a stacked area plot
ax = dataFrame.plot(kind='area', stacked=True, cmap='viridis', alpha=0.7)
plot.title('Stacked Area Plot: Peak Temperature For Vancouver and Victoria in all seasons')
plot.ylabel('Peak Temperature')
plot.xlabel('Seasons')
plot.show()
```

# # Example1- Percentage-Based Area Plot:

```python
# Normalize data to percentages
df_percentage = dataFrame.div(dataFrame.sum(axis=1), axis=0) * 100

# Draw a percentage-based area plot
ax = df_percentage.plot(kind='area', cmap='viridis', alpha=0.7)
plt.title('Percentage-Based Area Plot: Peak Temperature For Vancouver and Victoria in all seasons')
plt.ylabel('Percentage of Peak Temperature')
plt.xlabel('Seasons')
plt.show()
```

# Example2 – #Percentage plot


```python

 #DataFrame representing the sales of different products over several years. 
 #We will use an area plot to visualize the contribution of each product to the total sales over time.

gdpRevenue = {"Food" :(200, 192, 193),

              "Sports"       :(495, 475, 488),

              "Electronics" :(400, 418, 431),

              "Clothing":(200, 220, 230),

              "Others"            :(150, 155, 170)

             }            

index    = ("2010", "2011", "2012")
           

df   = pd.DataFrame(data=gdpRevenue)

normalized  = df.div(df.sum(axis=1), axis=0)*100

normalized.index = index

print(normalized)
#print(dataFrame.sum(axis=1))# means add coloumns togeather so for each row we have one value
#print(dataFrame)



#Percentage based
normalized.plot.area(stacked=True,cmap='viridis', alpha=0.7)
plt.title('Percentage-Based Area Plot:contribution of each product to the total sales over time')
plt.ylabel('Percentage of sales')
plt.xlabel('years')
plot.show()

```

# Example2 – Overlapping Area Plot


```python
# Overlapping Area Plot
normalized.plot.area(stacked=False, cmap='viridis', alpha=0.7)
plot.title('Overlapping Area Plot: Contribution of Each Product to Total Sales Over Time')
plot.ylabel('Percentage of Sales')
plot.xlabel('Years')
```

# Example2 – Stacked Area Plot


```python

# Stacked Area Plot
normalized.plot.area(stacked=True, cmap='viridis', alpha=0.7)
plot.title('Stacked Area Plot: Contribution of Each Product to Total Sales Over Time')
plot.ylabel('Percentage of Sales')
plot.xlabel('Years')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(x) + np.cos(x)

# Create DataFrame
df = pd.DataFrame({'X': x, 'Y1': y1, 'Y2': y2, 'Y3': y3})

# Overlapping Area Plot
plt.figure(figsize=(10, 4))
plt.fill_between(df['X'], df['Y1'], alpha=0.3, label='Y1')
plt.fill_between(df['X'], df['Y2'], alpha=0.3, label='Y2')
plt.fill_between(df['X'], df['Y3'], alpha=0.3, label='Y3')
plt.title('Overlapping Area Plot')
plt.legend()
plt.show()

# Stacked Area Plot
plt.figure(figsize=(10, 4))
plt.stackplot(df['X'], df['Y1'], df['Y2'], df['Y3'], labels=['Y1', 'Y2', 'Y3'], alpha=0.7)
plt.title('Stacked Area Plot')
plt.legend()
plt.show()

# Percentage-Based Area Plot
df_percentage = df[['Y1', 'Y2', 'Y3']].div(df[['Y1', 'Y2', 'Y3']].sum(axis=1), axis=0) * 100
plt.figure(figsize=(10, 4))
plt.fill_between(df['X'], df_percentage['Y1'], alpha=0.7, label='Y1 (%)')
plt.fill_between(df['X'], df_percentage['Y2'], alpha=0.7, label='Y2 (%)')
plt.fill_between(df['X'], df_percentage['Y3'], alpha=0.7, label='Y3 (%)')
plt.title('Percentage-Based Area Plot')
plt.legend()
plt.show()
```
